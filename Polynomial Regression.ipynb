```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
X = []
Y = []
for line in open('data_poly.csv'):
    x, y = line.split(',')
    x = float(x)
    X.append([1, x, x*x]) # add the bias term x0 = 1
    Y.append(float(y))
```

```python
X = np.array(X)
Y = np.array(Y)
```

```python
plt.scatter(X[:, 1], Y)
plt.show()
```

```python
# calculate weights
```

```python
w = np.linalg.solve(np.dot(X.T, X), np.dot(X.T, Y))
Yhat = np.dot(X, w)
```

```python
plt.scatter(X[:,1], Y)
plt.plot(sorted(X[:,1]), sorted(Yhat))
plt.show()
```

```python
# calculate r-squared
```

```python
d1 = Y - Yhat
d2 = Y - Y.mean()
r2 = 1 - d1.dot(d1) / d2.dot(d2)
print("the r-squared is:", r2)
```
